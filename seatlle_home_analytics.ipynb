```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Seattle Homes - Sheet1.tsv', sep='\t')
df.head()
```

```python
import seaborn as sb
```

```python
df.plot(x='Price/sq', y='Price', kind='scatter')
```

There is not much of a relation between price and price/sq. This is partially explained by the fact that price is highly dependent on the location of the property. Next we try to see if there is a relation between size of the property and price/sq.

```python
df.plot(x='Price/sq', y='Sq', kind='scatter')
```

```python
plt.figure(figsize=(20, 10))
sb.swarmplot(x="Area", y="Price", data=df)
```

```python
plt.figure(figsize=(20, 8))
sb.swarmplot(x="Area", y="Commute (Amazon)", data=df)
```
